```python
import pandas as pd
import numpy
import xlsxwriter
df=pd.read_excel('D:\AffinArabic\DataSet\AabicToEnglishDataset.xlsx')
```

```python
df
```

```python
sentences=df['English']
actual_class=df['Class']
ArabicSentences=df['Arabic']
```

```python
from pattern.en import sentiment
predicted_class=[]
workbook = xlsxwriter.Workbook('D:\AffinArabic\output\ArabicSentimentAnalysisUsing-pattern.xlsx')
worksheet1 = workbook.add_worksheet()
row=0

worksheet1.write(row, 0, str("Arabic Sentence"))
worksheet1.write(row, 1, str("English Sentence"))
worksheet1.write(row, 2, str("Actual Class"))
worksheet1.write(row, 3, str("Predicted Score"))
worksheet1.write(row, 4, str("Predicted Class"))
i=0
for text in sentences:
    analysis_result = sentiment(text)

    # Extract sentiment polarity and subjectivity
    sentiment_polarity, sentiment_subjectivity = analysis_result

    # Determine sentiment label based on polarity
    if sentiment_polarity >= 0:
        sentiment_label = 'Positive'
        predicted_class.append('Positive')
    else:
        sentiment_label = 'Negative'
        predicted_class.append('Negative')


    # Output results
    print(f"Text: {text}")
    print(f"Sentiment Polarity: {sentiment_polarity}")
    print(f"Sentiment Subjectivity: {sentiment_subjectivity}")
    print(f"Sentiment Label: {sentiment_label}")
    row=row+1
    worksheet1.write(row, 0, str(ArabicSentences[i]))
    worksheet1.write(row, 1, str(text))
    worksheet1.write(row, 2, str(actual_class[i]))
    worksheet1.write(row, 3, str(sentiment_polarity))
    worksheet1.write(row, 4, str(sentiment_label))
    i=i+1
workbook.close()  
 
    
```

```python
predicted_class
```

```python
actual_class
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(predicted_class,actual_class)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(predicted_class,actual_class)
```

```python
from sklearn.metrics import classification_report
print(classification_report(predicted_class,actual_class))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.heatmap(cm, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
import matplotlib.pyplot as plt
TN=110
FP=119
TP=210
FN=10

# Calculate metrics
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

# Create a bar graph
labels = ['TN', 'FP', 'TP', 'FN']
values = [TN, FP, TP, FN]

plt.bar(labels, values, color=['pink', 'brown', 'yellow', 'green'])
plt.xlabel('Confusion Matrix Based on pattern.en')
plt.ylabel('Count')
plt.title('Confusion Matrix')
plt.show()

# Print additional metrics
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1_score:.2f}')
```
