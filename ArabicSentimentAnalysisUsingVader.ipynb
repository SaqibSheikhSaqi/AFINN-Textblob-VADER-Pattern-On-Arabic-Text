```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import xlsxwriter

# Download the VADER lexicon (if not already downloaded)
#nltk.download('vader_lexicon')

# Create a SentimentIntensityAnalyzer object
sid = SentimentIntensityAnalyzer()

```

```python
import pandas as pd
import numpy
df=pd.read_excel('D:\AffinArabic\DataSet\AabicToEnglishDataset.xlsx')
```

```python
df
```

```python
sentences=df['English']
actual_class=df['Class']
ArabicSentences=df['Arabic']
```

```python
predicted_class=[]
workbook = xlsxwriter.Workbook('D:\AffinArabic\output\ArabicSentimentAnalysisUsing-VADER.xlsx')
worksheet1 = workbook.add_worksheet()
row=0

worksheet1.write(row, 0, str("Arabic Sentence"))
worksheet1.write(row, 1, str("English Sentence"))
worksheet1.write(row, 2, str("Actual Class"))
worksheet1.write(row, 3, str("Positive Score"))
worksheet1.write(row, 4, str("Negative Score"))
worksheet1.write(row, 5, str("Predicted Class"))
i=0
for sentence in sentences:
    sentiment_scores = sid.polarity_scores(sentence)
    print(f"Sentence: {sentence}")
    print(f"Sentiment Scores: {sentiment_scores}")

    # Determine sentiment based on the compound score
    if sentiment_scores['pos'] > sentiment_scores['neg']:
        print("Sentiment: Positive")
        sentiment='Positive'
        predicted_class.append('Positive')
    else:
        print("Sentiment: Negative")
        predicted_class.append('Negative')
        sentiment='Negative'

    row=row+1
    worksheet1.write(row, 0, str(ArabicSentences[i]))
    worksheet1.write(row, 1, str(sentence))
    worksheet1.write(row, 2, str(actual_class[i]))
    worksheet1.write(row, 3, str(sentiment_scores['pos']))
    worksheet1.write(row, 4, str(sentiment_scores['neg']))
    worksheet1.write(row, 5, str(sentiment))
    i=i+1
workbook.close()  
 
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(predicted_class,actual_class)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(predicted_class,actual_class)
```

```python
from sklearn.metrics import classification_report
print(classification_report(predicted_class,actual_class))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.heatmap(cm, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
